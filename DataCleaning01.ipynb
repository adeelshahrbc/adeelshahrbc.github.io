```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/adeelshahrbc/adeelshahrbc.github.io/master/MessyData.csv')

df.head()
df.info()
```

```python
df.describe(include ='all')
```

```python
df['City']=df['City'].str.replace('Ottawas$','Ottawa', regex=True)

df['City']=df['City'].str.replace('^Toronot$','^Toronto$', regex=True)
```
